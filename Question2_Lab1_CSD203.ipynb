```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def addToHead(self, x):
        new_node = Node(x)
        new_node.next = self.head
        self.head = new_node

    def addToTail(self, x):
        new_node = Node(x)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def addAfter(self, p, x):
        temp = self.head
        while temp:
            if temp.data == p:
                new_node = Node(x)
                new_node.next = temp.next
                temp.next = new_node
                return
            temp = temp.next
        print(f"Node with value {p} not found.")

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=' -> ')
            temp = temp.next
        print('None')

    def deleteFromHead(self):
        if self.head:
            deleted_node = self.head
            self.head = self.head.next
            return deleted_node.data
        return None

    def deleteFromTail(self):
        if not self.head:
            return None
        if not self.head.next:
            return self.deleteFromHead()
        temp = self.head
        while temp.next.next:
            temp = temp.next
        deleted_node = temp.next
        temp.next = None
        return deleted_node.data

    def deleteAfter(self, p):
        temp = self.head
        while temp and temp.next:
            if temp.data == p:
                deleted_node = temp.next
                temp.next = temp.next.next
                return deleted_node.data
            temp = temp.next
        return None

    def del_(self, x):
        if self.head and self.head.data == x:
            return self.deleteFromHead()
        temp = self.head
        while temp and temp.next:
            if temp.next.data == x:
                deleted_node = temp.next
                temp.next = temp.next.next
                return deleted_node.data
            temp = temp.next
        return None

    def search(self, x):
        temp = self.head
        while temp:
            if temp.data == x:
                return temp
            temp = temp.next
        return None

    def count(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count
```

```python
sll = SinglyLinkedList()
sll.addToHead("apple")
sll.addToTail("banana")
sll.addAfter("apple", "cherry")
sll.traverse() 
print("Count of nodes:", sll.count())  
print("Deleted from head:", sll.deleteFromHead())  
sll.traverse() 
print("Deleted from tail:", sll.deleteFromTail())  
sll.traverse()  
sll.addAfter("cherry", "date")
sll.traverse() 
print("Deleted after 'cherry':", sll.deleteAfter("cherry"))  
sll.traverse()  
print("Searching for 'date':", "Found" if sll.search("date") else "Not found")
print("Deleting 'date':", "Deleted" if sll.del_("date") else "Not deleted")
sll.traverse()  
```
